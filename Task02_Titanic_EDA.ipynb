```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
url = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/main/Task%202/train.csv"
df = pd.read_csv(url)
```

```python

# Drop duplicates
df = df.drop_duplicates()

# Handle missing values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

if 'Cabin' in df.columns:
    df = df.drop(columns=['Cabin'])
```

```python
plt.figure(figsize=(8,6))

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df, palette='Set2')
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df, palette='Set3')
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True, color="blue")
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df, palette='husl')
plt.title("Survival by Embarked Port")
plt.show()
```

# 📝 Insights & Conclusion

### 1. Data Cleaning  
- Missing values in **Age** were filled with the median.  
- Missing values in **Embarked** were filled with the mode.  
- The **Cabin** column was dropped due to too many missing values.  
- Duplicate records were removed.  

---

### 2. Key Findings from EDA  
- **Gender Effect:** Female passengers had a significantly higher survival rate compared to males.  
- **Class Effect:** First-class passengers had a much higher chance of survival than second and third class.  
- **Age Factor:** Younger passengers (especially children) had higher survival rates compared to older age groups.  
- **Fare:** Higher fare is positively correlated with survival, showing that wealthier passengers had better chances.  
- **Embarkation Port:** Passengers who boarded at port ‘C’ had a higher survival rate compared to ‘S’ and ‘Q’.  

---

### 3. Conclusion  
- Survival on the Titanic was **not random**. It was strongly influenced by **gender, passenger class, age, and fare paid**.  
- The famous phrase “**women and children first**” is supported by the data.  
- Wealth and social status (indicated by fare and class) also played a crucial role in survival chances.
